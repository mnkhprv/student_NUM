```python
import math
import re
from collections import Counter
from pathlib import Path
```

```python
class NaiveBayesSpamClassifier:
    def __init__(self, alpha=1.0):
        # alpha нь Laplace smoothing-ийн параметр.
        self.alpha = alpha
        self.vocabulary = []
        self.word_to_index = {}
        self.train_matrix = []
        self.train_labels = []
        self.class_priors = {}
        self.class_word_totals = {}
        self.conditional_probs = {}
        self.default_unknown_prob = {}
        self.is_trained = False

    def tokenize(self, text):
        """Текстийг жижиг үсэгт оруулж unigram token-ууд болгон задална."""
        return re.findall(r"\b[a-z0-9']+\b", text.lower())

    def build_vocabulary(self, documents):
        """Бүх сургалтын баримтаас давтагдахгүй үгсийн сан үүсгэнэ."""
        vocab = sorted({word for doc in documents for word in doc})
        self.vocabulary = vocab
        self.word_to_index = {word: index for index, word in enumerate(vocab)}

    def vectorize_documents(self, documents):
        """Bag-of-words матриц үүсгэнэ.

        Мөр бүр нь нэг захиа, багана бүр нь нэг ялгаатай үг байна.
        Нүд бүрт тухайн үг хэдэн удаа орсныг хадгална.
        """
        matrix = []

        for doc in documents:
            row = [0] * len(self.vocabulary)
            word_counts = Counter(doc)

            for word, count in word_counts.items():
                if word in self.word_to_index:
                    row[self.word_to_index[word]] = count

            matrix.append(row)
        return matrix

    def train(self, spam_texts, ham_texts):
        """Spam болон ham сургалтын өгөгдлөөр Multinomial Naive Bayes сургана."""
        spam_documents = [self.tokenize(text) for text in spam_texts]
        ham_documents = [self.tokenize(text) for text in ham_texts]

        all_documents = spam_documents + ham_documents

        if not all_documents:
            raise ValueError("Сургалтын өгөгдөл хоосон байна. DATA_DIR зам зөв эсэхийг шалгана уу.")

        self.train_labels = ["spam"] * len(spam_documents) + ["ham"] * len(ham_documents)

        # 1. Давтагдахгүй үгсийн сан үүсгэнэ.
        self.build_vocabulary(all_documents)

        # 2. Захиа x Үг хэмжээтэй bag-of-words матриц үүсгэнэ.
        self.train_matrix = self.vectorize_documents(all_documents)

        total_documents = len(all_documents)
        spam_document_count = len(spam_documents)
        ham_document_count = len(ham_documents)

        # P(spam), P(ham) урьдчилсан магадлалууд.
        self.class_priors = {
            "spam": spam_document_count / total_documents,
            "ham": ham_document_count / total_documents,
        }

        # Ангилал тус бүрийн бүх үгийн нийт давтамжийг тоолно.
        spam_word_counts = [0] * len(self.vocabulary)
        ham_word_counts = [0] * len(self.vocabulary)

        for row, label in zip(self.train_matrix, self.train_labels):
            if label == "spam":
                spam_word_counts = [current + value for current, value in zip(spam_word_counts, row)]
            else:
                ham_word_counts = [current + value for current, value in zip(ham_word_counts, row)]

        n_spam = sum(spam_word_counts)
        n_ham = sum(ham_word_counts)
        n_vocab = len(self.vocabulary)

        self.class_word_totals = {
            "spam": n_spam,
            "ham": n_ham,
            "vocab": n_vocab,
        }

        self.conditional_probs = {"spam": {}, "ham": {}}

        # P(w_i | spam), P(w_i | ham)-ийг Laplaceace smoothing-тэй тооцно.
        for index, word in enumerate(self.vocabulary):
            self.conditional_probs["spam"][word] = (spam_word_counts[index] + self.alpha) / (n_spam + self.alpha * n_vocab)
            self.conditional_probs["ham"][word] = (ham_word_counts[index] + self.alpha) / (n_ham + self.alpha * n_vocab)

        # Сургалтын үед үзэгдээгүй үг таарахад ашиглах default магадлал.
        self.default_unknown_prob = {
            "spam": self.alpha / (n_spam + self.alpha * n_vocab),
            "ham": self.alpha / (n_ham + self.alpha * n_vocab),
        }

        self.is_trained = True

    def predict_log_scores(self, text):
        """Захианы spam, ham log-score-уудыг буцаана."""
        if not self.is_trained:
            raise ValueError("Эхлээд train() функцийг ажиллуулна уу.")

        tokens = self.tokenize(text)
        token_counts = Counter(tokens)

        spam_score = math.log(self.class_priors["spam"])
        ham_score = math.log(self.class_priors["ham"])

        # Олон магадлал үржүүлэхийн оронд log авч нийлбэрээр бодно.
        for word, count in token_counts.items():
            spam_prob = self.conditional_probs["spam"].get(word, self.default_unknown_prob["spam"])
            ham_prob = self.conditional_probs["ham"].get(word, self.default_unknown_prob["ham"])

            spam_score += count * math.log(spam_prob)
            ham_score += count * math.log(ham_prob)

        return {"spam": spam_score, "ham": ham_score}

    def predict(self, text):
        """Илүү өндөр score-той ангиллыг буцаана."""
        scores = self.predict_log_scores(text)
        return "spam" if scores["spam"] > scores["ham"] else "ham"

    def predict_many(self, texts):
        return [self.predict(text) for text in texts]

    def evaluate(self, spam_texts, ham_texts):
        """Dev set дээр accuracy болон confusion matrix тооцно."""
        predictions = []
        gold_labels = []

        for text in spam_texts:
            predictions.append(self.predict(text))
            gold_labels.append("spam")

        for text in ham_texts:
            predictions.append(self.predict(text))
            gold_labels.append("ham")

        correct = sum(pred == gold for pred, gold in zip(predictions, gold_labels))
        total = len(gold_labels)

        confusion = {
            "tp": 0,
            "tn": 0,
            "fp": 0,
            "fn": 0,
        }

        for pred, gold in zip(predictions, gold_labels):
            if gold == "spam" and pred == "spam":
                confusion["tp"] += 1
            elif gold == "ham" and pred == "ham":
                confusion["tn"] += 1
            elif gold == "ham" and pred == "spam":
                confusion["fp"] += 1
            else:
                confusion["fn"] += 1

        return {
            "accuracy": correct / total,
            "correct": correct,
            "total": total,
            "confusion_matrix": confusion,
        }
```

```python
def resolve_data_dir():
    """Notebook-ийг хаанаас ажиллуулснаас үл хамааран spam_data хавтсыг олно."""
    cwd = Path.cwd().resolve()
    direct_candidates = [cwd / "Laboratory8" / "spam_data", cwd / "spam_data"]

    for candidate in direct_candidates:
        if (candidate / "train" / "spam").exists() and (candidate / "train" / "ham").exists():
            return candidate

    for parent in [cwd, *cwd.parents]:
        if parent.name == "spam_data" and (parent / "train" / "spam").exists() and (parent / "train" / "ham").exists():
            return parent

        for candidate in parent.rglob("spam_data"):
            if (candidate / "train" / "spam").exists() and (candidate / "train" / "ham").exists():
                return candidate

    raise FileNotFoundError("spam_data хавтас олдсонгүй. Project-ийн зөв хавтсаас notebook-ийг ажиллуулж байгаа эсэхээ шалгана уу.")


def read_emails_from_directory(directory_path):
    """Өгөгдсөн хавтас дахь бүх текст файлыг уншиж жагсаалтаар буцаана."""
    texts = []

    for file_path in sorted(Path(directory_path).glob("*.txt")):
        texts.append(file_path.read_text(encoding="latin-1"))

    return texts


DATA_DIR = resolve_data_dir()
print(f"Өгөгдлийн хавтас: {DATA_DIR}")

train_spam = read_emails_from_directory(DATA_DIR / "train" / "spam")
train_ham = read_emails_from_directory(DATA_DIR / "train" / "ham")
dev_spam = read_emails_from_directory(DATA_DIR / "dev" / "spam")
dev_ham = read_emails_from_directory(DATA_DIR / "dev" / "ham")

print(f"Train spam захиа: {len(train_spam)}")
print(f"Train ham захиа:  {len(train_ham)}")
print(f"Dev spam захиа:   {len(dev_spam)}")
print(f"Dev ham захиа:    {len(dev_ham)}")
```

```python
model = NaiveBayesSpamClassifier(alpha=1.0)
model.train(train_spam, train_ham)

print("Bag-of-words матрицын хэмжээ:")
print(f"  Мөрийн тоо    = {len(model.train_matrix)}  (нийт сургалтын захиа)")
print(f"  Баганын тоо   = {len(model.vocabulary)}  (нийт ялгаатай үг)")
print()
print("Naive Bayes параметрүүд:")
print(f"  P(spam)       = {model.class_priors['spam']:.6f}")
print(f"  P(ham)        = {model.class_priors['ham']:.6f}")
print(f"  N_spam        = {model.class_word_totals['spam']}")
print(f"  N_ham         = {model.class_word_totals['ham']}")
print(f"  N_vocab       = {model.class_word_totals['vocab']}")

results = model.evaluate(dev_spam, dev_ham)
print()
print("Dev set үнэлгээ:")
print(f"  Accuracy      = {results['accuracy']:.4f} ({results['correct']}/{results['total']})")
print(f"  TP            = {results['confusion_matrix']['tp']}")
print(f"  TN            = {results['confusion_matrix']['tn']}")
print(f"  FP            = {results['confusion_matrix']['fp']}")
print(f"  FN            = {results['confusion_matrix']['fn']}")
```

```python

example_text = dev_spam[0]
scores = model.predict_log_scores(example_text)

print("Жишээ захианы таамаглал:")
print(f"  Таамагласан ангилал = {model.predict(example_text)}")
print(f"  spam log-score      = {scores['spam']:.4f}")
print(f"  ham log-score       = {scores['ham']:.4f}")
```
