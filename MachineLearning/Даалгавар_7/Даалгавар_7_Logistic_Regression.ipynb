# Даалгавар 7: Логистик регресс


# 7 Хэрэгжүүлэлт (10 оноо)

Энэ даалгаврын зорилго нь **хоёр ангилалт логистик регресс** ашиглан ажиллах чадвартай **хэл боловсруулалтын (NLP) систем** хэрэгжүүлэх явдал юм. Таны алгоритм зоогийн газрын сэтгэгдэл **эерэг** эсвэл **сөрөг** эсэхийг тодорхойлно.

**Тэмдэглэл:** Хэрэгжүүлэлт хэсгийг эхлэхээсээ өмнө, энэхүү даалгаварт хэрэг болох чухал ойлголтуудыг эхэлж сайн унших хэрэгтэй.

---

## 7.1 Даалгавар

### Өгөгдөл

Энэ даалгаварт шаардлагатай өгөгдлийг агуулсан zip файлыг татаж авна уу. Эдгээр өгөгдөл нь **Yelp dataset**-аас авсан болно. Өгөгдлийн файлуудын мөр бүр нэг жишээ өгөгдөл бөгөөд дараах зүйлсээс бүрдэнэ:
- **шошго** (`0` бол сөрөг сэтгэгдэл, `1` бол эерэг сэтгэгдэл)
- **үгсийн багц**

Жишээ бүрийн (өөрөөр хэлбэл мөр бүрийн) формат нь дараах байдалтай байна:

`label\tword1 word2 word3 ... wordN\n`

Энд:
- үгс нь хоорондоо **хоосон зайгаар** тусгаарлагдана,
- шошго нь үгсээс **таб тэмдэгтээр** тусгаарлагдана.

Өгөгдлийн жишээ:

```
1 i will never forget this single breakfast experience in mad...
0 the search for decent chinese takeout in madison continues ...
0 sorry but me julio fell way below the standard even for med...
1 so this is the kind of food that will kill you so there s t...
```

### Онцлогийн инженерчлэл (Feature Engineering)

Лекц дээр бид логистик регрессийг **тогтмол урттай бодит утгат оролтод** (жишээ нь $x^{(i)} \in \mathbb{R}^n$) хэрхэн хэрэглэхийг үзсэн. Гэвч энд сэтгэгдэл бүр:
- хувьсах урттай,
- бодит утгат вектор биш байна.

Тиймээс логистик регресс ажиллуулахын тулд өгөгдлийг эхлээд **онцлогийн инженерчлэлийн** суурь аргаар хувиргах шаардлагатай. Энэ даалгаварт бид **үгний эмбеддинг (word embeddings)** загварыг ашиглана. Дараагийн хэсэгт (7.2) үүнийг дэлгэрэнгүй тайлбарласан.

### Хэрэглэгдэх Код

Энэ даалгаварт ерөнхийдөө та хоёр программ бичнэ:
- `feature.py`
- `lr.py`

`feature.py` нь түүхий оролтын өгөгдлийг авч, сургалт, баталгаажуулалт, туршилтын жишээ бүрт **бодит утгат вектор** үүсгэнэ.  
`lr.py` нь эдгээр векторуудыг авч, жишээ бүр **эерэг** эсвэл **сөрөг** сэтгэгдэл эсэхийг таамаглахын тулд **логистик регрессийн загвар** сургана.

## 7.2 Онцлогийн загвар

Үгсийн багцыг вектор болгон хувиргахын тулд бид түгээмэл хэрэглэгддэг онцлогийн инженерчлэлийн арга болох **үгний эмбеддинг**-г ашиглана.

Энд:
- $\omega$ нь шинж тэмдгийн инженерчлэлийн аргыг,
- $x^{(i)}$ нь сургалтын жишээг (англи үгсийн багц) тус тус илэрхийлнэ.

### Үгний эмбеддинг санаа

Зөвхөн ямар үг орсон эсэхийг тэмдэглэхийн оронд, үгний эмбеддинг нь үг бүрийг **бага хэмжээст вектор орон зайд** дүрслэн, тухайн үгийн **утга зүйн мэдээллийг** агуулж чаддаг.

Энэ даалгаварт бид **GloVe embeddings**-ийг ашиглана. Энэ нь өргөн хэрэглэгддэг үгний векторын багц юм.

### Embeddings

`glove_embeddings.txt` файлд **6792 үгийн GloVe embedding** агуулагдана. Сэтгэгдэл дэх бүх үгс өгөгдсөн `glove_embeddings.txt` файлд байх албагүй. Ийм байхгүй үгсийг **үгсийн сангаас гадуурх (out-of-vocabulary)** гэж үзээд **үл тоомсорлоно**.

Мөр бүр дараах бүтэцтэй байна:

`word\tfeature1\tfeature2\t...feature300\n`

Өөрөөр хэлбэл үг бүрийн embedding нь **300 хэмжээст вектор** байна.

Жишээ нь, `glove_embeddings.txt` файлын эхний хэдэн мөр (утгуудыг 3 орны нарийвчлалтай дугуйруулсан):

```
deserves  0.175  -0.153  -0.208  0.092  0.222  0.202  ...
butter    0.357   0.469  -0.021  0.024 -0.168 -0.213  ...
staffing -0.076   0.212  -0.384  0.552 -0.193 -0.052  ...
weird     0.110   0.090   0.139  0.340 -0.098 -0.113  ...
```

## Үгний эмбеддинг ашиглах нь

Энэ загварт шинж тэмдгийн инженерчлэл **хоёр алхамтай** байна:

### 1-р алхам: GloVe тольд байхгүй үгсийг хасах

Сэтгэгдлийн доторх `glove_embeddings.txt` файлд **байгаа** үгсийг л авч үлдэнэ.

$$x_{\text{trim}}^{(i)} = \text{TRIM}\bigl(x^{(i)}\bigr)$$

Энд $\text{TRIM}(x^{(i)})$ нь $x^{(i)}$-ийн үгсээс зөвхөн `glove_embeddings.txt` дотор байгаа үгсийг авч үлдсэн хувилбар юм.

### 2-р алхам: Үлдсэн үгсийн GloVe embedding-үүдийн дунджийг авах

Дараа нь тайрсан хувилбар $x_{\text{trim}}^{(i)}$-ийг эцсийн шинж тэмдгийн вектор болгохын тулд түүний үгсийн GloVe embedding-үүдийн **дундаж**-ийг авна:

$$\omega\bigl(x^{(i)}\bigr) = \frac{1}{J}\sum_{j=1}^{J} \text{GloVe}\bigl(x_{\text{trim},j}^{(i)}\bigr)$$

Энд:
- $J$ нь $x_{\text{trim}}^{(i)}$-д үлдсэн үгсийн тоо,
- $x_{\text{trim},j}^{(i)}$ нь $x_{\text{trim}}^{(i)}$-ийн $j$-р үг,
- $\text{GloVe}\bigl(x_{\text{trim},j}^{(i)}\bigr) \in \mathbb{R}^{300}$ нь тухайн үгийн 300 хэмжээст GloVe вектор юм.

---

### Жишээ

$x^{(i)}$ өгүүлбэр нь: **"a hot dog is not a sandwich because it is not square"**

Жишээ GloVe толь нь:

```
hot        0.1   0.2   0.3
not       -0.1   0.2  -0.3
sandwich   0.0  -0.2   0.4
square     0.2  -0.1   0.5
```

Тэгвэл $x_{\text{trim}}^{(i)}$ нь: **"hot not sandwich not square"**

Энд GloVe тольд байхгүй үгсийг хассан. Мөн үгийн дарааллыг хадгалж, давхардсан үгсийг устгахгүй.

Иймээс $x^{(i)}$-ийн шинж тэмдгийг дараах байдлаар тооцно:

$$
\omega\bigl(x^{(i)}\bigr)
= \frac{1}{5}\Bigl(
  \text{GloVe}(\textit{hot})
  + 2\cdot\text{GloVe}(\textit{not})
  + \text{GloVe}(\textit{sandwich})
  + \text{GloVe}(\textit{square})
\Bigr)
= \begin{bmatrix} 0.02 \\ 0.06 \\ 0.12 \end{bmatrix}
$$

**Тайлбар:** Давхардсан үгсийг хадгалах нь тухайн үгийг **давтамжаар нь жинлэхтэй адил**. Жишээлбэл, `"good"` гэдэг үг `"bad"`-аас 3 дахин олон давтагдвал тухайн сэтгэгдэл эерэг байх магадлал өндөр байна.

## 7.3 `feature.py`

`feature.py` нь дээр тайлбарласан **үгний шигтгээний аргыг** хэрэгжүүлж, түүхий сургалтын жишээг (шошго + англи үгсийн жагсаалт) форматлагдсан сургалтын жишээ (шошго + шинж тэмдгийн вектор) болгон хувиргана.

### Оролт

- Сургалт, баталгаажуулалт, туршилтын оролтын өгөгдөл
- Шинж тэмдэг гарган авахад ашиглах **GloVe embeddings**

### Гаралт

- Сургалт, баталгаажуулалт, туршилтын **форматлагдсан өгөгдөл**
- Та сургалт, баталгаажуулалт, туршилтын **бүх багц дээр** feature extraction хийх ёстой.

### Гаралтын формат

Гаралтын файл бүр жишээ бүрийн форматлагдсан хувилбарыг **шинэ мөрөөр** хэвлэсэн байна. Шинэ мөр үүсгэхдээ `\n` ашиглана.

Мөр бүрийн формат дараахтай **яг адил** байна:

`label\tvalue1\tvalue2\tvalue3\t...valueM\n`

Энд:
- мөр бүр нэг зоогийн газрын сэтгэгдэлд харгалзана,
- эхний утга нь **шошго**,
- үлдсэн утгууд нь **шинж тэмдгийн векторын элементүүд** байна.

Бүх элементүүд **таб тэмдэгтээр** тусгаарлагдана. Гаралтаа **таслалаас хойш 6 орны нарийвчлалтай** дугуйруулна уу.

Жишээ:

```
1.000000  -0.166646   0.641027  -0.064805  ...
0.000000  -0.224874   0.461526  -0.215232  ...
0.000000  -0.222178   0.437475  -0.083073  ...
1.000000  -0.215923   0.612535   0.061671  ...
```

## 7.4 `lr.py`

`lr.py` нь форматлагдсан өгөгдлийг авч, тухайн зоогийн газрын сэтгэгдэл **сөрөг** эсвэл **эерэг** эсэхэд харгалзах **0 эсвэл 1 шошго** гаргах **логистик регрессийн ангилагч**-ийг хэрэгжүүлнэ.

### Оролт

- Сургалт, баталгаажуулалт, туршилтын форматлагдсан өгөгдөл (`feature.py`-ийн гаргасан файлууд)
- Сургах **epoch**-ийн тоо — командын мөрийн аргументаар дамжин орж ирнэ
- **Суралцах хурд (learning rate)** — мөн командын мөрийн аргументаар дамжин орж ирнэ

### Шаардлагууд

- Загвартаа **intercept term** (огтлолцлын гишүүн)-ийг оруулна. Үүнийг тусдаа хувьсагч гэж авч үзэж болно, эсвэл параметрийн векторт нэгтгэж болно (**зөвлөмжтэй арга**). Аль ч тохиолдолд intercept параметрийг зөв шинэчлэх ёстой.
- Загварын бүх параметрийг **0-ээр эхлүүлнэ**.
- Логистик регрессийг сургахдаа **stochastic gradient descent (SGD)** ашиглана.
- SGD шинэчлэлтийг сургалтын өгөгдөлд **файлд өгөгдсөн дарааллаар нь** хийнэ.
- Ер нь SGD-д сургалтын жишээнүүдийг холих нь түгээмэл байдаг ч, энэ даалгаврыг **autograde** хийхийн тулд сургалт **deterministic** байх ёстой. Иймээс **сургалтын өгөгдлийг бүү shuffle хий**.

## Гаралт

- Сургалтын болон туршилтын өгөгдлийн **таамагласан шошгууд**
- Сургалтын болон туршилтын **алдааны хэмжүүрүүд**

### Гаралтын шошгуудын формат

`lr` програм нь сургалтын өгөгдөл болон туршилтын өгөгдөл дээрх загварын таамаглалыг агуулсан **хоёр `.txt` файл** үүсгэнэ. Файл бүр жишээ бүрийн таамагласан шошгыг **шинэ мөрөөр** агуулна. Эдгээр файлын нэрийг командын мөрийн аргументаар өгнө. Шинэ мөр үүсгэхдээ `\n` ашиглана.

Жишээ:

```
1
0
0
1
```

### Гаралтын хэмжүүрийн формат

Таны програм сургалт дууссаны дараах **сургалтын** болон **туршилтын алдааг** тайлагнах нэг `.txt` файл үүсгэнэ. Файлын нэр командын мөрийн аргументаар өгөгдөнө.

Таны тайлагнасан бүх тоон утга **лавлах шийдлээс 0.00001-оос ихгүй зөрүүтэй** байх ёстой бөгөөд алдааны утгуудыг **таслалаас хойш 6 орны нарийвчлалтай** дугуйруулна.

Жижиг өгөгдлийн сан дээр, **500 epoch**, **0.1 learning rate**-тай сургасны дараах лавлах шийдлийн жишээ:

```
error(train): 0.000000
error(test): 0.625000
```

Файл дахь мөр бүр `\n` шинэ мөрийн тэмдэгтээр төгссөн байна. Мөн **colon (`:`)-ы дараа нэг хоосон зай** байна.
